```python
from doctr.io import DocumentFile
from doctr.models import ocr_predictor

import os
import torch
import time
import re
```

```python
model = ocr_predictor(reco_arch='parseq',pretrained=True,assume_straight_pages=False)
```

```python
image_file = "/teamspace/studios/this_studio/page_image/1wMr9ofP.jpg"

#load image file
single_img_doc = DocumentFile.from_images(image_file)

#perform OCR given the defined model object from above
result = model(single_img_doc)

#show the OCR results
print(result.render())
```
