```python
from PIL import Image
import numpy as np
from surya.recognition import RecognitionPredictor
from surya.detection import DetectionPredictor
```

```python
detection_predictor = DetectionPredictor()
recognition_predictor = RecognitionPredictor()
```

```python
image_file = "images/Screenshot 2026-06-02 223503.png"

#open the image
image = Image.open(image_file)

#Perform OCR: Call the recognition model with the given image, the language, and the detection model to use
predictions = recognition_predictor([image], det_predictor=detection_predictor)

#Put the results of each line into one string
surya_ocr_str = ""

for line in predictions[0].text_lines:
    surya_ocr_str = surya_ocr_str + ' ' + line.text
```

```python
print(surya_ocr_str)
```
