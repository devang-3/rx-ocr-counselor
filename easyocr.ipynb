```python
import easyocr
import cv2

import matplotlib.pyplot as plt

from glob import glob
from random import sample
from PIL import ImageFont, ImageDraw, Image
```

```python
reader = easyocr.Reader(['en', 'tr'], gpu=False)
```

```python
img= cv2.imread('/teamspace/studios/this_studio/images/Screenshot 2026-06-02 223454.png')
img2= cv2.imread('/teamspace/studios/this_studio/images/Screenshot 2026-06-02 224238.png')
img3= cv2.imread('/teamspace/studios/this_studio/images/Screenshot 2026-06-02 224210.png')
```

```python
result = reader.readtext(img, detail =1, paragraph =True)
result2 = reader.readtext(img2, detail =1, paragraph =True)
result3 = reader.readtext(img3, detail =1, paragraph =True)
```
